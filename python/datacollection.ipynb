```python
import yfinance as yf
import pandas as pd

stocks=["AAPL","MSFT","GOOG","TSLA","JPM","PFE","KO","XOM","NVDA","AMZN","^GSPC"]
data=yf.download(stocks,start="2020-01-01")
```

```python
data.to_csv('stock_data.csv')
print('Data saved to stock_data.csv')
```

```python
import pandas as pd
data=pd.read_csv("stock_data.csv",header=[0,1],index_col=0)
# The KeyError 'Adj Close' indicates that 'Adj Close' is not in the MultiIndex columns.
# This is because yfinance.download with auto_adjust=True (default) removes the 'Adj Close' column.
# Correcting to access 'Close' from the first level (level=0) of the MultiIndex:
prices = data.xs('Close', level=0, axis=1)
prices = prices.ffill()
prices.to_csv("clean_prices.csv")
```

```python
weight = 1/len("Stocks")
portfolio=pd.DataFrame({
    'Stocks':stocks,
    'weight':[weight]*len(stocks)
})
portfolio.to_csv('portfolio.csv')
```
